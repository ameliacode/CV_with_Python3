## 📸 컴퓨터 비전과 머신러닝의 통합
###  Why Machine Laerning?
이전 장에서 이미지 분할 시, 이미지와 물체와 배경을 분리하기 위해 여러 알고리즘(윤곽선 검출, 슈퍼 픽셀, 그래프 컷)을 사용하면서 분할을 해보았다. 이미지 분할 시, 파라미터를 수동해서 조절했는데, 머신러닝을 통해 수월하게 진행할 수 있는 장점을 지니기에 많은 연구들이 머신러닝과 결합하여 진행해왔다.  
기본적으로 머신러닝은 이미지 특성들을 학습한 후 알려지지 않은 이미지(테스트 이미지)를 입력으로 labeling을 한다.

#### 데이터 전처리
데이터의 다양성을 최대한 확보하는 과정이다. 여러 변화에 대응할 수 있도록 선행적으로 처리하는 과정.  
* 이미지 이동
* 이미지 회전/크기 조정

### Logistic Regression 

````python
from sklearn import datasets, metrics
from sklearn.linear_model import LogisticRegression
from skimage import io, color, feature, transform
from sklearn.preprocessing import StandardScaler
````

````python
mnist = datasets.load_digits()
images = mnist.images
data_size = len(images)
````

````python
images = images.reshape(len(images),-1)  #Data Preprocessing → LosisticRegression 함수는 1D입력이다.
labels = mnist.target
````

````python
LR_classifier = LogisticRegression(C=0.01, penalty="l2", tol=0.01)
LR_classifier.fit(images[:int(data_size * 0.75)], labels[:int(data_size * 0.75)]) #for training 75%
````

````python
predictions = LR_classifier.predict(images[int(data_size * 0.75):])  #for testing 25%
target = labels[int(data_size*0.75): ]
````

````python
print("Performance Report: \n %s \n" %(metrics.classification_report(target,predictions)))
````

````python
digit_img = io.imread("digit_6.png")
digit_img = color.rgb2gray(digit_img)
digit_img = transform.resize(digit_img, (8,8), mode="wrap")
digit_edge = feature.canny(digit_img, sigma=5)
digit_edge = [digit_edge.flatten()]
prediction = LR_classifier.predict(digit_edge)
````

````python
io.imshow(digit_img)
io.show()
print(prediction[0])
````

### Support Vector Machine
데이터를 영역으로 나누는 초평면을 찾는다. 선형 회귀와 다르게 선형으로 분리되지 않는 데이터까지 분류할 수 있다.

````python
from sklearn import datasets, metrics, svm
````

````python
mnist = datasets.load_digits()
images = mnist.images
data_size = len(images)
````

````python
images = images.reshape(len(images),-1)
labels = mnist.target
````

````python
SVM_classifier = svm.SVC(gamma=0.001)
SVM_classifier.fit(images[:int(data_size * 0.75)], labels[:int(data_size*0.75)]) #for training 75%
````

````python
predictions = SVM_classifier.predict(images[int(data_size * 0.75):]) #for testing 25%
target = labels[int(data_size * 0.75):]
````

````python
print("Performance Report: \n %s \n" %(metrics.classification_report(target, predictions)))
````

#### 주성분 분석(Principle Component Analysis)

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, decomposition
````

````python
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure()
    ax = plt.subplots(figsize=(8,5))
    
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(digits.target[i]),
                 color=plt.cm.Set1(Y[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
````

````python
digits = datasets.load_digits(n_class=6)
X = digits.data
Y = digits.target
````

````python
n_samples, n_features = X.shape
n_neighbors = 30
X_pca = decomposition.TruncatedSVD(n_components=2).fit_transform(X)
````

````python
plot_embedding(X_pca)
````

````python
from sklearn import manifold
````

````python
tsne = manifold.TSNE(n_components=2, init="pca", random_state=0)
X_tsne = tsne.fit_transform(X)
````

````python
plot_embedding(X_tsne)
````

### K-mean Clustering
비지도 학습의 한 종류, 유사도 척도를 기반으로 서로 가까운 점들끼리 클러스터를 이룬다.

````python
from sklearn import datasets, metrics
from sklearn.cluster import KMeans
````

````python
mnist = datasets.load_digits()
images = mnist.images
data_size = len(images)
````

````python
images = images.reshape(len(images),-1)
labels = mnist.target
````

````python
clustering = KMeans(n_clusters=10, init="k-means++", n_init=10) ## k 중심점 초기화 by lyod 알고리즘
clustering.fit(images[:int(data_size * 0.75)])
````

````python
print(clustering.labels_)
predictions = clustering.predict(images[int(data_size*0.75):])
````

````python
print(predictions)
````
